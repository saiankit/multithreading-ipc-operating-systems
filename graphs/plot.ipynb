```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set_style("whitegrid")
sns.set_palette('husl' ,3) 
```

```python
df = pd.read_csv('p1_10000.csv')
```

```python
 data = pd.DataFrame({ 'Threads': df['Threads'],
                      'Time(seconds)': df['Time']});
```

```python
sns.lineplot(x="Threads", y="Time(seconds)",
             data=data)
  
plt.show()
```
